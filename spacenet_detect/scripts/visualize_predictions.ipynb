```python
import os
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io
```

```python
savepath = "../checkpoints"
files = os.listdir(savepath)
files
```

```python
lbls = np.load(os.path.join(savepath, files[0]))
preds = np.load(os.path.join(savepath, files[1]))
```

```python
preds.shape
```

```python
fig, ax = plt.subplots(1,2, figsize = (10,5))
ax[0].imshow(preds[1][0])
ax[1].imshow(lbls[1][0])
```

```python
for i in range(11):
    io.imsave(f"preds_{i}.png", preds[i][0])
    io.imsave(f"lbls_{i}.png", lbls[i][0])
```
