```python
import os
import torch
import pandas as pd
from skimage import io, transform
import numpy as np
import matplotlib.pyplot as plt
import random
from torch.utils.data import Dataset, DataLoader, RandomSampler
from torchvision import transforms, utils
from tools.data_utils import PinDataset, Resize, ToTensor
```

```python
data_folder = "./data/Synthetic/28by28pixels_1000images_123456seed"
d1 = 28
d2 = 28
```

```python
grid_sparse_pins_file = f"{data_folder}/mesh_10step_pins/pins.csv"
grid_dense_pins_file = f"{data_folder}/mesh_3step_pins/pins.csv"
random_dense_pins_file = f"{data_folder}/random_100pins/pins.csv"
random_sparse_pins_file = f"{data_folder}/random_10pins/pins.csv"
```

```python
def show_pins(image, pins, outputs):
    """Show image with pins"""
    plt.imshow(image, extent=[0,d1,0,d2])
    plt.scatter(pins[:, 0], pins[:, 1], s=24, marker='o', c='r')
    # plt.pause(0.001)  # pause a bit so that plots are updated
    # there is a shift in the pins probably

# plt.figure()
# show_pins(io.imread(f"{data_folder}/images/{img_name}"), pins, outputs)
# plt.show()
```

```python
ind = random.randint(0, 999)
```

```python
grid_sparse_dataset = PinDataset(csv_file=grid_sparse_pins_file,
                             root_dir=f"{data_folder}/images/",
                             transform=transforms.Compose([ToTensor(),
                                                           Resize()]))

sample = grid_sparse_dataset[ind]
sample['image'] = sample['image'].permute(1, 2, 0)

fig = plt.figure()
plt.tight_layout()
plt.axis('off')
show_pins(**sample)
fig.savefig(f'example_images/synth_grid_sparse.pdf', bbox_inches='tight', pad_inches=0)
```
